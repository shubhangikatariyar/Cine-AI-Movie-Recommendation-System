## Convert Genres data

```python
movies.iloc[0].genres
```

```python
#INPUT: '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

#OUTPUT: ['Action', 'Adventure', 'Fantasy', 'ScienceFiction']
```

```python
import ast

def convert(x: list) -> list:
    genres_list=[]
    if type(x)==str:
        x = ast.literal_eval(x)

    for single_dict in x:
        genres_list.append(single_dict["name"])

    return genres_list
    
        
    
```

```python
convert([{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}])
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
import ast
def convert_cast_top3(x: str)->list:
    top3_Cast=[]
    if type(x)==str:
        x = ast.literal_eval(x)
    for single_dict in x[:3]:
        top3_Cast.append(single_dict['name'])
    return top3_Cast


    
```

```python
movies['cast']=movies['cast'].apply(convert_cast_top3)
```

## Convert crew data

```python
movies.crew[0]
```

```python
# FIND name where job: 'director'

import ast

def find_director(x: str) -> list:
    director=[]
    if type(x)==str:
        x = ast.literal_eval(x)

    for single_dic in x:
        if single_dic["job"].lower()=='director':
            director.append(single_dic["name"])
            break
    return director
            
        
```

```python
find_director(movies.crew[0])
```

```python
movies['crew']=movies['crew'].apply(find_director)
```

# STEMMING

Why is there a need for STEMMING?  

There are words like abuse, abusive, aceepted, accepts 

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text:str)-> str:
    y = []
    for i in text.split():
        y.append(ps.stem(i)) # stemming every word

    return " ".join(y)
```

```python
new_df['tags'][0]
```

```python
stem(new_df['tags'][0])
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# COSINE SIMILARITY

EUCLIDEAN DISTANCE:  tip to tip distance between the vectors  [FAILURE IN HIGH DIMENSIONAL FEATURE]  
COSINE DISTANCE: angle between two vectors

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors).shape # 4806 movies ka 4806 movies k saath distance
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[0] # PEHLI MOVIE KA HAR MOVIE K SAATH DISTANCE  hence sabse pehla wala is 1
```

```python
similarity[1] # 2nd MOVIE KA HAR MOVIE K SAATH DISTANCE  hence sabse dusra wala is 1
```

```python
# the position index is very important in this case: Hence we retain it
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6] # FIRST 5 movies with high cosine similarity
```

```python
def recommend(movie):
    '''
    movie: input
    return: 5 similar movies suggestions
    '''
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6] 

    for i in movies_list:
        #print(i[0])
        print(new_df.iloc[i[0]].title)
    return None
```

```python
recommend('Avatar')
```
